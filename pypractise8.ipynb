### Categorical Plots

Seaborn helps un in doing the analytics on Categorical Data points . In this section we will discuss about 
- boxplot
- violinplot
- countplot
- bar plot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
## Count plot

sns.countplot('sex', data=df )
```

```python
sns.countplot(y='sex', data=df)
```

```python
sns.countplot('smoker', data=df )
```

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(data=df, 'day')
```

### BAR PLOT

```python
# its similar as count plot but we need to specify the x-axis and y-axis
sns.barplot( x= 'total_bill', y= 'sex', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

### BOX PLOT

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('sex', 'total_bill', data=df)
```

```python
sns.boxplot('day', 'total_bill', data=df )
```

```python
sns.boxplot('day', 'total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x = 'total_bill', y='day', hue='smoker', data=df)
```

### VIOLIN PLOT

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot('total_bill', 'day', data=df, palette='rainbow')
```
